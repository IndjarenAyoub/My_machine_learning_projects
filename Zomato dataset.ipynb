```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

```python
plt.bar(country_names[:3], country_val[:3])
plt.xlabel('Country')
plt.ylabel('Values')
plt.title('Top 3 Countries')
plt.show()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
# Plotting the bar chart
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.xlabel('Number of Ratings')
plt.ylabel('The reatings')
#plt.title('Top 10 Most Prolific Directors on Netflix')
plt.show()
```

```python
palette=['white','red','orange','yellow','green','green']
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color' , data=ratings, palette=palette)
```

#Observations
1. Most records are Not rated
2. Then maximum number of ratings are between 2.5 and 3.4

```python
sns.set_style('whitegrid')
sns.countplot(x='Rating color',data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

```python
rating_counts = df['Rating text'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette="viridis")
plt.xlabel('Rating text')
plt.ylabel('Number of occurences')
plt.title('Ranking of ratings by occurences')
plt.show()
```
